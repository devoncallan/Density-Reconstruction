## Reconstructing 2D Electron Density Maps

###### Set Reconstruction parameters.

```python
group   = 'p6mm'
F_coeff = [3.968, 10, 7.108, 2.409, 1.959, 1.0283]
step = 0.005
recon_set = ReconstructionSet(F_coeff, step=step)
args = recon_set.get_args()
```

###### Generate reconstructions for each phase.
phase_list stores all phases for a given number of peaks (6), including inverse phases (e.g ++++++ AND ------).
<br>
Reconstruction is constructed for each phase and added to the reconstruction set.
<br>
All reconstructions normalized by the absolute maximum value of all reconstructions in the set.

```python
phase_list = recon_set.enumerate_phases(num_peaks=6,
                                        inv_phases=True)

for phase in phase_list:
    recon = Reconstruction(phase=phase, **args) 
    recon.calculate_density()
    recon_set.add_reconstruction(recon)
recon_set.normalize_all()
```

###### Plot electron density map of a given reconstruction.

```python
phase_str = '+--+++'
Plot.plot_2D_density(recon_set.reconstructions[phase_str],
                    figsize=(3, 3), dpi=150,
                    cmap=cmap)
```

## Defining Electron Density Models

###### Set model parameters
Initialize volume fractions and electron densities for triblock copolymer model

```python
f_P4MCL = 0.13
f_PTFEA = 0.51
f_PDDA  = 0.36

p_P4MCL = 0.342 # e- density of P4MCL (e-/Å^3)
p_PTFEA = 0.436 # e- density of PTFEA (e-/Å^3)
p_PDDA  = 0.316 # e- density of PDDA  (e-/Å^3)

p_avg = f_P4MCL*p_P4MCL + f_PTFEA*p_PTFEA + f_PDDA*p_PDDA

p_dict = {f_P4MCL: p_P4MCL,
          f_PTFEA: p_PTFEA,
          f_PDDA:  p_PDDA}
```

###### Generate core-shell and nested lattice models. 
Save model object in model_dict.

```python
f_dict_models = {'CS1': (f_PTFEA, f_PDDA, f_P4MCL),
                 'CS2': (f_PDDA, f_PTFEA, f_P4MCL),
                 'CS3': (f_P4MCL, f_PDDA, f_PTFEA),
                 'CS4': (f_P4MCL, f_PTFEA, f_PDDA),
                 'CS5': (f_PDDA, f_P4MCL, f_PTFEA),
                 'CS6': (f_PTFEA, f_P4MCL, f_PDDA),
                 'NL1': (f_PDDA, f_PTFEA, f_P4MCL),
                 'NL2': (f_PTFEA, f_PDDA, f_P4MCL),
                 'NL3': (f_P4MCL, f_PTFEA, f_PDDA),
                 'NL4': (f_PTFEA, f_P4MCL, f_PDDA),
                 'NL5': (f_P4MCL, f_PDDA, f_PTFEA),
                 'NL6': (f_PDDA, f_P4MCL, f_PTFEA)}

model_dict = {}

for model_str in f_dict_models:
    f = f_dict_models[model_str]
    model = Model(model_str, f, p_dict, step=step)
    model_dict[model_str] = model
```

###### Plot electron density map of a given model.

```python
model_str = 'CS4'
Plot.plot_2D_density(model_dict[model_str],
                    figsize=(3, 3), dpi=150,
                    cmap=cmap)
```

## Fitting each reconstruction to each model

```python
recon_dict = {} # Reconstruction for each model/reconstruction pair
SSR_dict = {}   # Residual sum of squares (SSR)
R2_dict = {}    # Coefeficient of determination (R^2)
K_dict = {}     # K (scaling const.) dictionary

# Loop through all possible reconstructions
for phase_str in recon_set.reconstructions:
    print(phase_str)
    recon_ref = recon_set.reconstructions[phase_str]
    
    # Loop through all models
    for model_str in model_dict:
        
        model = model_dict[model_str]
        
        fit_recon, K, error_params = recon_ref.fit_to_model(model)
        
        model_phase_str = model_str + '_' + phase_str
        recon_dict[model_phase_str] = fit_recon
        K_dict[model_phase_str]     = K
        SSR_dict[model_phase_str] = error_params[0]
        R2_dict[model_phase_str] = error_params[1]
```

###### Plot $R^2$ for each model-reconstruction pair in descending order

N: # of models-reconstruction pairs to plot
<br>
highlight_model: name of model to highlight

```python
R2_sorted_dict = sorted(R2_dict.items(), key=lambda item: item[1], reverse=True)
Plot.plot_R2_histogram(R2_sorted_dict, 
                       N=50, 
                       highlight_model='CS4', 
                       title='')
```

###### Plot x and y slice of a given fitted electron density map and corresponding model

```python
phase_str = '+--+++'
model_str = 'CS4'
model_phase_str = model_str + '_' + phase_str

y_val = 1
Plot.plot_1D_density(recon=recon_dict[model_phase_str],
                     model=model_dict[model_str], 
                     y_val=y_val,
                     title=model_phase_str,
                     dpi=150,
                     ylim=[0.28, 0.48])
```

```python
x_val = SQRT3/2
Plot.plot_1D_density(recon=recon_dict[model_phase_str],
                     model=model_dict[model_str], 
                     x_val=x_val,
                     title=model_phase_str, 
                     dpi=150,
                     ylim=[0.28, 0.48])
```
